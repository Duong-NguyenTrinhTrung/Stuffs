```python
import os 
import numpy as  np
path="drive/MyDrive/Dong-Data"
```

```python

print([i for i in range(2008, 2020)])
N=42600
for i in range(10):
  with open(path+"/Table2_SampleDistributionByYear"+str(i)+".csv","w") as f:
    f.write("Year,Frequency,Percentage\n")
    data=pd.read_csv(path+"/FinalExtract"+str(i)+".csv",  encoding = 'unicode_escape')
    print("file ",i)
    # print(data.columns)
    # print(data.shape)
    # print(data.head(5))
    nrowsList=[]
    nrowsListPercents=[]
    for year in range(2008, 2020):
      nrows=len(data[data["YEAR"]==year])
      nrowsList.append(nrows)
      nrowsListPercents.append('{:.2f}'.format(100*float(nrows)/N))
      f.write(str(year)+","+str(nrows)+","+'{:.2f}'.format(100*float(nrows)/N)+"\n")
    f.write("Grand Total,"+str(N)+",100.0%")
    print(nrowsList)
    print(nrowsListPercents)
  
```
