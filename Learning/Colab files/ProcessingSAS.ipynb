```python
path="drive/MyDrive/Dong-Data"
```

```python
from sas7bdat import SAS7BDAT
with SAS7BDAT(path+"/SAS/Whiz.sas7bdat") as file:
  permo=file.to_data_frame()
print(type(permo))
print(permo.shape)
```

```python
length=int(len(permo)/100)
for part in range(99):
  permo.iloc[part*length:(part+1)*length].to_csv(path+"/SAS/permo-part"+str(part)+".csv", header=True, index=False)

permo.iloc[99*length:].to_csv(path+"/SAS/permo-part99.csv", header=True, index=False)
```
