```python
import os 
import numpy as  np
path="drive/MyDrive/Dong-Data"
```

```python
def change_ticker_value(x):
	return x[x.find(":")+1:]
```

```python
data=pd.read_excel(path+"/Compustast-1990-2019/1990.xls")
# print(columnName)
columnName=data.iloc[6].tolist()
columnName[1]="TICKER"
print(columnName)
```

```python
CompustastData=pd.DataFrame()
filenames=os.listdir(path+"/Compustast-1990-2019")
# print(filenames)
#duyet qua tat ca cac field trong thu muc CompustastData
for i in range(len(filenames)):
    filename=filenames[i]
    #doc file
    data=pd.read_excel(path+"/Compustast-1990-2019/"+filename)
    #gan ten cot cua data bang columnName
    data.columns=columnName
    #doi ten cho cot Total assets thanh ten ngan hon
    columnName[15]="Total Assets"
    columnName[17]="Total Common Equity"
    columnName[18]="Market Capitalization"
    data.columns=columnName
    #chi lay tu dong sat dong tieu de tro di
    data=data.loc[7:,:]
    #thay the gia tri "-" bang NaN tren toan dataframe
    data.replace("-",np.nan, inplace=True)
    #tinh mean cua rieng cot "Total assets"
    meanTotalAsset=np.mean(data['Total Assets'])
    #thay gia tri NaN cua cot Total asset bang mean cua cot do
    data['Total Assets'].fillna(meanTotalAsset, inplace=True)
    #tinh va them cot "size" bang log tu nhien cua total asset
    data['size']=np.log(data['Total Assets'])
    
    #tinh mean cua rieng cot "Total Common Equity"
    meanTotalCommonEquity=np.mean(data['Total Common Equity'])
    #thay gia tri NaN cua cot Total Common Equity bang mean cua cot do
    data['Total Common Equity'].fillna(meanTotalCommonEquity, inplace=True)
    
    #tinh mean cua rieng cot "Total Common Equity"
    meanMarketCapitalization=np.mean(data['Market Capitalization'])
    #thay gia tri NaN cua cot Total Common Equity bang mean cua cot do
    data['Market Capitalization'].fillna(meanMarketCapitalization, inplace=True)
    #tinh va them cot "Growth" bang 
    data['Growth']=(data['Total Assets']-data['Total Common Equity']+data['Market Capitalization'])/data['Total Assets']
    #tinh lai gia tri cua TICKER
    data["TICKER"]= data["TICKER"].apply(change_ticker_value)
    
    #dua data vao CompustastData
    print(i," data shape ",data.shape)
    if i==0:
      CompustastData=data
    else:
      CompustastData=CompustastData.append(data)
    print(i, " CompustastData.shape ",CompustastData.shape)
    
```
